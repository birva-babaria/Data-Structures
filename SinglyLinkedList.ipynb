```python
# In singly linked list node contains: Key and Pointer
```

```python
class Node:

  #executed when node is created
  def __init__(self, data = None, next = None):
    self.data = data
    self.next = next

  #executed when type of object is called
  def __str__(self):
    return "Data = {} and Next = {}".format(self.data, self.next)

class SinglyLinkedList:

  #executed when new Linked List is created
  def __init__(self):
    self.head = None
    self.tail = None

  #executed when type of object is called
  def __str__(self):
    return "New Singly Linked List is created!!"

  #computing length of single LL
  def length(self):
    itr = self.head
    count = 0
    while(itr):
      count += 1 
      itr = itr.Next
    return count

  #printing singly linked list
  def printSLL(self):
    if self.head == None:
      return "Can not print empty list"
    itr = self.head
    lst = ""
    while(itr):
      lst += ' --> ' + str(itr.data)
      itr = itr.next
    return lst
  
  # -----OPERATION ON SINGLY LINKED LIST-----

  #Insert a new node at the front
  def pushFront(self, data):
    node = Node(data, self.head)
    if self.head == None and self.tail == None:
      self.tail = node
    self.head = node

  #Remove a node from the front
  def popFront(self):
    if self.head == None and self.tail == None:
      print("Can not pop from empty list")
      return
    if self.head == self.tail:
      self.head = self.tail = None
      return
    self.head = self.head.next
    return

  #Insert a new node at the end
  def pushBack(self, data):
    node = Node(data, None)
    if self.head == None and self.tail == None:
      self.head = self.tail = node
      return
    self.tail.next = node
    self.tail = node
  
  #Remove a node from the back
  def popBack(self):
    if self.head == None and self.tail == None:
      print("Can not pop from empty list")
      return
    if self.head == self.tail:
      self.head = self.tail = None
      return
    itr = self.head
    while(itr):
      if itr.next == self.tail:
        itr.next = None
        self.tail = itr
        return
      itr = itr.next

  #Find a node and return boolean value
  def Find(self, data):
    if self.head == None:
      return False
    itr = self.head
    while(itr):
      if itr.data == data:
        return True
      itr = itr.next
    return False

  #Check if LL is empty or not (bool)
  def isEmpty(self):
    if self.head == None:
      return True
    return False

  #Erase the node using data as an argument
  def eraseUsingKey(self, data):
    if not self.Find(data):
      return
    #if list is empty
    if self.head == None and self.tail == None:
      raise Exception("Can not erase from empty linked list")
    #if list has one element
    if self.head == self.tail:
      if self.head.data == data:
        self.head = self.tail = None
        return
    itr = self.head
    while(itr):
      #id key is first element
      if itr.data == data:
        self.head = itr.next
        return
      if itr.next.data == data:
        itr.next = itr.next.next
        #if key is last element
        if itr.next == None:
          self.tail = itr
        return
      itr = itr.next    
```
