```python
#supporting class
class Queue:
  def __init__(self):
    self.queue = []

  def enqueue(self, value):
    self.queue.append(value)

  def dequeue(self):
    l = len(self.queue)
    if l == 0:
      return -1
    value = self.queue[0]
    del self.queue[0]
    return value

  def isEmpty(self):
    if len(self.queue) == 0:
      return True
    return False

  def length(self):
    return len(self.queue)

  def printQueue(self):
    return self.queue
```

```python
class TreeNode:

  #executed when new object is created
  def __init__(self, data):
    self.left = None
    self.right = None
    self.data = data

  #executed when the type of the pbject is called
  def __str__(self):
    return "New node is created"
```

```python
#inserting a node in the BST
def insertNode(root, data):
  if data <= root.data:
    if root.left is None:
      root.left = TreeNode(data)
    else:
      insertNode(root.left, data)
  elif data > root.data:
    if root.right is None:
      root.right = TreeNode(data)
    else:
      insertNode(root.right, data)

#deleting a node from BST
def deleteNode(root, data):
  if root is None:
    return None
  if data < root.data:
    root.left = deleteNode(root.left, data)
  elif data > root.data:
    root.right = deleteNode(root.right, data)
  else:
    # if node is the leaf node
    if root.left is None and root.right is None:
      root = None
    #if node has one child
    elif root.left is None:
      root = root.right
    elif root.right is None:
      root = root.left
    #if node has 2 childrens
    else:
      successor = findMinNode(root.right)
      root.data = successor.data
      root.right = deleteNode(root.right, successor.data)
  return root

#Find minimum element of the tree
def findMin(root):
  while root.left is not None:
    root = root.left
  return root.data

#find minimum node of the BST
def findMinNode(root):
  while root.left is not None:
    root = root.left
  return root

#Find maximum element of the tree
def findMax(root):
  while root.right is not None:
    root = root.right
  return root.data

#-----TREE TRAVERSAL------

# DFS
def printInorder(root):
  if root.left:
    printInorder(root.left)
  print(root.data, end = " ")
  if root.right:
    printInorder(root.right)

def printPreorder(root):
  print(root.data, end = " ")
  if root.left:
    printPreorder(root.left)
  if root.right:
    printPreorder(root.right)

def printPostorder(root):
  if root.left:
    printPostorder(root.left)
  if root.right:
    printPostorder(root.right)
  print(root.data, end = " ")

#BFS
def levelTraversal(root):
  queue = Queue()
  queue.enqueue(root)
  while not queue.isEmpty():
    node = queue.dequeue()
    print(node.data, end = " ")
    if node.left is not None:
      queue.enqueue(node.left)
    if node.right is not None:
      queue.enqueue(node.right)
```

```python
root = TreeNode(27)
insertNode(root, 14)
insertNode(root, 35)
insertNode(root, 10)
insertNode(root, 19)
insertNode(root, 31)
insertNode(root, 42)
```
