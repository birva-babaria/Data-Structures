```python
class Node():
  def __init__(self, data = None, next = None):
    self.data = data
    self.next = next

  def __str__(self):
    return "Data = {} || Next = {}".format(self.data, self.next)
  
class LinkedList():
  def __init__(self):
    self.head = None
  
  def __str__(self):
    return "New LinkedList is created!!"

  ### LENGTH OF THE LINKED LIST
  def length(self):
    itr = self.head
    count = 0
    while(itr):
      count += 1
      itr = itr.next
    return count

#----------------------------------------------------

  ### INSERT NEW NODE AT THE BEGINNING
  def insert_at_beginning(self, data):
    node = Node(data, self.head)
    self.head = node
  
  ### INSERT NEW NODE AT THE END
  def insert_at_end(self, data):
    node = Node(data, None)
    if self.head == None:
      node = Node(data, None)
      self.head = node
      return
    itr = self.head
    while(itr):
      if itr.next == None:
        break      
      itr = itr.next
    itr.next = node

  ### INSERT VALUES BY A LIST
  def insert_values(self, lst):
    self.head = None
    for l in lst:
      self.insert_at_end(l)

  ### INSERT VALUE BY INDEX
  def insert_at(self, data, index):
    if self.head == None:
      print("Linked List is empty!!")
      return
    if index<0 or index>self.length():
      raise Exception("Index out of bound")
    if index == 0:
      self.insert_at_beginning(data)
      return
    if index == self.length():
      self.insert_at_end(data)
      return
    count = 0
    itr = self.head
    tmp_itr = Node()
    while(itr):
      if count == index:
        new_node = Node()
        new_node.data = data
        new_node.next = itr
        tmp_itr.next = new_node
        return
      count += 1
      tmp_itr = itr
      itr = itr.next

#----------------------------------------------------
  ### REMOVE ELEMENT BY INDEX
  def remove_at(self, index):
    if self.head == None:
      print("Linked List is empty!!")
      return
    if index<0 or index>=self.length():
      raise Exception("Index out of bound")
    if index == 0:
      self.head = self.head.next
      return
    count = 0
    itr = self.head
    tmp_itr = Node()
    while(itr):
      if count == index:
        tmp_itr.next = itr.next
        return
      count += 1
      tmp_itr = itr
      itr = itr.next

#----------------------------------------------------

  ### PRINTING LINKED LIST
  def print_LinkedList(self):
    if self.head == None:
      print("Linked List is empty!!")
      return
    
    itr = self.head
    llstr = ""
    while(itr):
      llstr += " --> " + str(itr.data)
      itr = itr.next
    return llstr
  
  ### PRINTING NODE VALUE BY INDEX
  def print_by_index(self, index):
    if self.head == None:
      print("Linked List is empty!!")
      return
    old_index = index
    itr = self.head
    while(index >= 0):
      if itr == None:
        print("Err: Index Out Of Bound")
        return
      if index == 0:
        print("Element at index {} is = {}".format(old_index, itr.data))
      index -= 1
      itr = itr.next
      
```
