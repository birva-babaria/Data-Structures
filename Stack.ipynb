```python
class Stack:
  def __init__(self):
    self.stack = []
  
  def push(self, value):
    self.stack.append(value)
  
  def pop(self):
    l = len(self.stack)
    if l == 0:
      return -1
    value = self.stack[l-1]
    del self.stack[l-1]
    return value

  def isEmpty(self):
    if len(self.stack) == 0:
      return True
    return False
  
  def length(self):
    return len(self.stack)
  
  def printStack(self):
    return self.stack
```

```python
s1 = Stack()
s1.push(1)
print("Stack = ",s1.printStack())
s1.push(2)
print("Stack = ",s1.printStack())
print("length = ",s1.length())
print(s1.pop())
print("Stack = ",s1.printStack())
print(s1.isEmpty())
print(s1.pop())
print("Stack = ",s1.printStack())
print(s1.isEmpty())
```
